```python
#codigo feito na mão
while True:
  print("Escolha o operador:")
  print("+ para Soma")
  print("- para Subtração")
  print("* para Multiplicação")
  print("/ para Divisão")

  operacao = (input("digite a forma de operação: "))
  n1 = int(input("digite o primeiro numero: "))
  n2 = int(input("digite o segundo numero: "))

  if operacao == "+":
    resultado = n1 + n2
    print(f"O resultado de {n1} + {n2} = {resultado}")
  elif operacao == "-":
    resultado = n1 - n2
    print(f"O resultado de {n1} - {n2} = {resultado}")
  elif operacao == "*":
    resultado = n1 * n2
    print(f"O resultado de {n1} * {n2} = {resultado}")
  elif operacao == "/":
    resultado = n1 / n2
    print(f"O resultado de {n1} / {n2} = {resultado}")
  else:
    print("voce encerrou o programa")
    break
```

```python
#coisada pela Inteligencia artificial
# Define uma função para realizar as operações aritméticas
def calcular(n1, n2, operacao):
  """
  Realiza a operação aritmética especificada entre dois números.
  Retorna o resultado da operação ou uma mensagem de erro em caso de divisão por zero.
  """
  if operacao == "+":
    return n1 + n2
  elif operacao == "-":
    return n1 - n2
  elif operacao == "*":
    return n1 * n2
  elif operacao == "/":
    if n2 == 0:
      return "Erro: Não é possível dividir por zero."
    else:
      return n1 / n2

# Loop principal para o programa da calculadora
while True:
  # Exibe as operações disponíveis para o usuário
  print("\n--- Calculadora Simples ---")
  print("Escolha o operador:")
  print("+ para Soma")
  print("- para Subtração")
  print("* para Multiplicação")
  print("/ para Divisão")
  print("Ou digite 'sair' para encerrar o programa.")

  # Obtém a entrada da operação do usuário
  operacao = input("Digite a forma de operação (ou 'sair'): ").lower()

  # Verifica se o usuário deseja sair
  if operacao == "sair":
    print("Programa encerrado. Até mais!")
    break

  # Valida a operação escolhida
  if operacao not in ["+", "-", "*", "/"]:
    print("Operação inválida. Por favor, escolha +, -, *, / ou 'sair'.")
    continue # Reinicia o loop para pedir a entrada novamente

  # Obtém as entradas numéricas do usuário com tratamento de erro
  try:
    n1 = float(input("Digite o primeiro número: ")) # Usa float para aceitar números decimais
    n2 = float(input("Digite o segundo número: "))  # Usa float para aceitar números decimais
  except ValueError:
    print("Entrada inválida. Por favor, digite números válidos.")
    continue # Reinicia o loop se a entrada não for um número

  # Realiza o cálculo usando a função auxiliar
  resultado = calcular(n1, n2, operacao)

  # Exibe o resultado ou a mensagem de erro
  if isinstance(resultado, str) and "Erro" in resultado:
    print(resultado) # Imprime a mensagem de erro para divisão por zero
  else:
    # Formata a saída baseada no resultado ser inteiro ou float
    if resultado == int(resultado):
      # Se o resultado for um número inteiro, mostra sem casas decimais (se os inputs também forem inteiros)
      if n1 == int(n1) and n2 == int(n2):
        print(f"O resultado de {int(n1)} {operacao} {int(n2)} = {int(resultado)}")
      else:
        print(f"O resultado de {n1} {operacao} {n2} = {int(resultado)}")
    else:
      print(f"O resultado de {n1} {operacao} {n2} = {resultado}")
  print("----------------------------------------") # Separador para clareza
```

```python
#Missão 1
print("organizador de arquivos")


arquivos = ["relatorio.pdf", "foto.jpg", "planilha.xlsx", "curriculo.pdf", "imagem.png"]

for arquivo in arquivos:
    if "." in arquivo:
        extensao = arquivo.split(".")[-1]


        pasta = extensao.upper() + "_arquivos"

        print(f"Criando pasta '{pasta}' e movendo o arquivo '{arquivo}'")
    else:
        print(f"Arquivo sem extensão: {arquivo}")

print("todos os arquivos foram organizados")
```

```python
#missao 2
print("Consulta CEP")

ceps = {
    "01001000": "Rua da Consolação, São Paulo - SP",
    "20040030": "Avenida Rio Branco, Rio de Janeiro - RJ",
    "30130010": "Praça Sete de Setembro, Belo Horizonte - MG"
}

cep = input("Digite o CEP (8 números, ex: 01001000): ").strip()

if cep in ceps:
    print("Endereço encontrado:")
    print(ceps[cep])
else:
    print("CEP não encontrado ou inválido.")

print("Consulta finalizada!")
```
