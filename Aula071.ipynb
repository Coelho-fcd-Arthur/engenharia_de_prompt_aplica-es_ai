```python
#refatoração 2
def e_primo(n):
    if n <= 1:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

numero = 29
#refatoração 3
if e_primo(numero):
    print("É primo")
else:
    print("Não é primo")
```
