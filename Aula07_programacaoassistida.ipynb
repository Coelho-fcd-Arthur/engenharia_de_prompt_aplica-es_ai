```python
# Define as funções básicas para cada operação matemática
def somar(a, b):
    return a + b

def subtrair(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    # Garante que não haja divisão por zero
    if b == 0:
        return "Erro: Divisão por zero!"
    return a / b

# Exibe o título da calculadora
print("--- Calculadora Simples ---")

# Solicita ao usuário os dois números e a operação desejada
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
operacao = input("Escolha a operação (+, -, *, /): ")

# Mapeia as operações para as funções correspondentes usando um dicionário
operacoes = {
    "+": somar,
    "-": subtrair,
    "*": multiplicar,
    "/": dividir
}

# Verifica se a operação escolhida é válida e executa a função correspondente
if operacao in operacoes:
    # Chama a função armazenada no dicionário para a operação selecionada
    resultado = operacoes[operacao](num1, num2)
else:
    resultado = "Operação inválida!"

# Exibe o resultado da operação
print("O resultado é:", resultado)
```
